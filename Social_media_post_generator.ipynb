```python
import google.generativeai as genai
from google.colab import userdata

genai.configure(
    api_key=userdata.get("GEMINI_API_KEYS")
)
```

```python
# Social Media Post Generator

def generate_post(topic, platform, tone):

    system_prompt = (
        "You are an expert social media content writer. "
        "Write natural, engaging posts suitable for different social media platforms."
    )

    user_prompt = f"""
    Generate a social media post.

    Topic: {topic}
    Platform: {platform}
    Tone: {tone}

    LinkedIn:
    - Around 120 words
    - Professional
    - End with a question
    - 3-5 hashtags

    Twitter/X:
    - Maximum 280 characters
    - 2 hashtags

    Instagram:
    - 80-100 words
    - Friendly
    - Use emojis
    - 5-8 hashtags

    Facebook:
    - Around 100 words
    - Conversational
    - 2-4 hashtags

    Generate only the post.
    """

    model = genai.GenerativeModel(
        model_name="gemini-3.5-flash-lite",
        system_instruction=system_prompt
    )

    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature":1,
            "max_output_tokens":300
        }
    )

    return response.text
```

```python
def save_post(topic, platform, post):

    filename = platform + "_post.txt"

    with open(filename, "w") as file:
        file.write("Topic: " + topic + "\n")
        file.write("Platform: " + platform + "\n\n")
        file.write(post)

    print("Post saved successfully as", filename)
```

```python
topic = input("Enter Topic: ")

platform = input("Enter Platform (LinkedIn/Twitter/Instagram/Facebook): ")

tone = input("Enter Tone: ")

post = generate_post(topic, platform, tone)

print("\nGenerated Post\n")

print(post)

save_post(topic, platform, post)
```
